```python
from transformers import *
import tensorflow as tf
```

```python
model_version = 'allenai/scibert_scivocab_uncased'
do_lower_case = True
model = TFBertModel.from_pretrained(model_version, from_pt=True)
tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=do_lower_case)
```

```python
model.summary()
```

```python
import pandas as pd
```

```python
data = pd.read_json('ai_papers.json')
```

```python
abstracts = data['abstract'].values
```

```python
abstracts_encoded=[tokenizer.encode(l ,return_tensors='tf',max_length=200, pad_to_max_length=True) for l in abstracts]

batch_size=100
batched=[]
for i in range(int(len(abstracts_encoded)/batch_size)+1):
    b = abstracts_encoded[i*batch_size:(i+1)*batch_size]
    batched.append(tf.concat(b,axis=0))
```

```python
from tqdm import tqdm
embed =[]
for i in tqdm(range(len(batched))):
    e = model(batched[i])[0]
    mean_e = tf.reduce_mean(e, axis=1)
    embed.append(mean_e)
```

```python
len(embed)
```

```python
import pickle
```

```python
with open('embeddings.pkl', 'wb') as f:
    pickle.dump(embed, f)
```

```python
fileo = open('embeddings.pkl', 'rb')
datao = pickle.load(fileo)
fileo.close()
```

```python
embeddings = []

for tensor in datao:
    tmp = tensor.numpy()
    for elem in tmp:
        embeddings.append(elem)
    
```
